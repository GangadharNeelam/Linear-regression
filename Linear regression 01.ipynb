LINEAR REGRESSION
--

        REGRESSION
            --> Regression is the realtionship between two or more variables, where the change in one variable(s) is associated with change in another variable(s).
            
        TYPES OF REGRESSION
            --> Simple linear regression
            --> Multiple linear regression
            --> Polynomial linear regression

        APPLICATIONS
            --> Economical Growth
            --> Product price
            --> Housing sales
            --> Score prediction
            
Linear regression
--

    Linear regression is a statistical model used to predict the relationship between independent and dependent variables.
    
        Y = MX + C
        
        Where,
            y = Dependent variable(s)
            x = Independent variable(s)
            m = Slope of the line
            c = Co-efficient of line or intercept (Point where function crosses y-axis)

Evaluation
--

        R-SQUARED VALUE
        ---------------
        
                    --> It measures that how close the data are fitted to regression line
                    --> Distance of actual mean vs Distance of predicted mean
                    --> It is also known as co-efficient of determination



                        R^2	=	coefficient of determination
                        RSS	=	sum of squares of residuals
                        TSS	=	total sum of squares
                        
                    * Higher the value better the model (0 to 1)

        MEAN ABSOLUTE ERROR
        -------------------
        
                    --> It is the average distance between actual and predicted values





                    * Lower the value better the model

            MEAN SQUARED ERROR
            ------------------
            
                    --> Average of squared distance between actual and predicted values
                    --> Measures variance of residuals
            


                    * Lower the value better the model

            ROOT MEAN SQUARED ERROR
            -----------------------
            
                    --> Root of mean squared error
                    --> Measures standard deviation of residuals
            


                    * Lower the value better the model

## Declare training and testing data

```python
X = data.drop('Life expectancy', axis = 1)
y = data['Life expectancy']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Creating a model

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

## Predictions

```python
predictions = reg.predict(X_test)
predictions
```

## Evaluate Model

```python
from sklearn import metrics
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import explained_variance_score,mean_squared_error, mean_squared_log_error, mean_absolute_error

mae = mean_absolute_error(y_test, predictions) # Lower the values better the model (0 to infinity)
print('Mean Absolute Error :', mae)

accuracy =  r2_score(y_test, predictions).round(4) # Higher the value better the model (0 to 1)
print('R Squared :', accuracy)

mse = mean_squared_error(y_test, predictions) # Lower the value better the model(0 to infinity)
print('Mean Squared Error :', mse)

mse_log_error = mean_squared_log_error(y_test, predictions) # Lower the value better the model (0 to infinty)
print('Mean Squared Log Error :', mse_log_error)
```

### with cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import KFold
```

```python
reg = LinearRegression()
k_fold = KFold(n_splits=10, shuffle=True, random_state=1)
cross_validation_score = cross_val_score(reg, X_train, y_train, scoring='r2', cv = k_fold)
print(cross_validation_score)
```

#### Hyper parameters tuning

```python
from sklearn.model_selection import GridSearchCV
k_folds = KFold(n_splits=10, shuffle=True, random_state=1)

hyper_parameters = [{'n_features_to_select':list(range(1, 25))}]

reg = LinearRegression()
reg.fit(X_train, y_train)
reg = RFE(reg)

model = GridSearchCV(estimator=reg, param_grid=hyper_parameters, verbose=1, cv=k_folds,
                    scoring='r2', return_train_score=True)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_estimator_
```

```python
model.best_score_
```

```python
reg.get_params()
```

```python
results= pd.DataFrame(model.cv_results_)
results
```

```python
plt.plot(results['param_n_features_to_select'], results['mean_test_score'])
plt.plot(results['param_n_features_to_select'], results['mean_train_score'])
plt.xlabel('n_features')
plt.ylabel('r squared')
plt.legend(['train', 'test'], loc = 'lower right')
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)

final_model = RFE(reg, n_features_to_select=16)
final_model.fit(X_train, y_train)

y_preds = final_model.predict(X_test)
score = r2_score(y_test, y_preds)
print('R Squared :', score)

mse = mean_squared_error(y_test, y_preds)
print('Mean Squared Error :', mse)

ms_log_e  = mean_squared_log_error(y_test, y_preds)
print('Mean Squared Log Error :', ms_log_e)
```
